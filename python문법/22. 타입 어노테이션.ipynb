ë³€ìˆ˜ê°€ íŠ¹ì • íƒ€ìž…ì´ê±°ë‚˜ Noneì¼ ìˆ˜ ìžˆëŠ” ê²½ìš° Optionalì„ ì‚¬ìš©í•©ë‹ˆë‹¤.

```python
def Hello(name: str | None) -> str:
    if name is None:
        return "Hello, Guest!"
    return f"Hello, {name}!"

print(Hello("apple"))
print(Hello(None))
# print(Hello())  //Hello() missing 1 required positional argument: 'name'
```

```python
def find_user(user_id: int) -> str | None:
    if user_id == 1:
        return "apple"
    return None

result1 = find_user(1)
result2 = find_user(99)

print(result1)
print(result2)
```

```python
users: dict[int, str] = {
    1: "apple",
    2: "banana"
}

def get_user(user_id: int) -> str | None:
    return users.get(user_id)
    # return users.get(user_id, "ì•Œ ìˆ˜ ì—†ìŒ")
```

```python
print(get_user(1))
print(get_user(2))
print(get_user(100))
```
